1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a data visualization library in Python that provides plotting functions and tools for creating static, animated, and interactive visualizations. It is widely used for generating high-quality plots, charts, and figures in various formats.

Line plot: A line plot is a basic plot that represents data points connected by straight line segments. It is useful for visualizing trends or patterns over time or any continuous variable.

Bar plot: A bar plot, also known as a bar chart, represents categorical data using rectangular bars. It is commonly used to compare values between different categories.

Scatter plot: A scatter plot displays individual data points as dots in two-dimensional space. It is useful for visualizing the relationship between two continuous variables and identifying any patterns or correlations.

Histogram: A histogram is a graphical representation that organizes a set of data into bins and displays the frequency or count of data points falling into each bin. It is commonly used to understand the distribution of a continuous variable.

Pie chart: A pie chart is a circular chart that represents data as slices of a pie, where each slice corresponds to a specific category. It is useful for showing the proportion or percentage of each category within a dataset.

2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
#A scatter plot is a type of plot that displays the relationship between two variables by representing individual data points as dots on a two-dimensional plane. It is particularly useful for visualizing the correlation or pattern between two continuous variables.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

3. Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()

plt.show()
```

4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()

plt.barh(company,profit,)
plt.show()
```

5. What is a box plot? Why is it used? Using the following data plot a box plot.

```python
#A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum values, along with any potential outliers.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.show()
```
