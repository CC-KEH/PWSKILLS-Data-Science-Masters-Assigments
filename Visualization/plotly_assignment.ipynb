1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px
titanic = sns.load_dataset('titanic')
px.scatter(data_frame=titanic,x='age',y='fare')
```

2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = px.data.tips()
px.box(data_frame= tips,x='day',y='total_bill')
```

3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
px.histogram(data_frame=tips,x='sex',y='total_bill',pattern_shape='smoker',color='day')
```

4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
iris = sns.load_dataset('iris')
px.scatter_matrix(data_frame=iris,color='species')
```

5. What is Distplot? Using Plotly express, plot a distplot.

```python
#In python distplot is deprecated, plotly now provides histogram function for creating, distplot like plot, by kernel density estimation in histogram function
sns.distplot(tips["total_bill"], kde=True, rug=True)
```
